# 텐서 데이터 유형

-대부분 텐서는 GPU 컴퓨터에 사용됨 <br>
-일반적인 기본 값: torch.float32 또는 torch.float   --> 32비트 부동 소수점 <br>
-16비트 부동 소수점(torch.float16 또는 torch.half)과 64비트 부동소수점(torch.float64 또는 torch.double)도 존재<br>
-8비트, 16비트, 32비트, 64비트 정수도 존재 <br>

-정밀도 값(8, 16, 32)이 높을수록 숫자를 표현하는 데 더 많은 세부 정보와 데이터가 사용됨<br>
-정밀도가 낮은 데이터 유형은 일반적으로 계산 속도가 빠르지만 정확도와 같은 평가 지표에서 성능이 떨어짐(계산 속도는 <br>빠르지만 정확도는 떨어짐)


- 특정 데이터 타입을 가진 텐서 생성

```python
float_32_tensor = torch.tensor([3.0, 6.0, 9.0],
                                dtype=None, #기본값은 None. torch.float32 또는 전달된 데이터 유형 사용
                                device=None, #기본값은 None. 기본 텐서 유형 사용
                                requires_grad= False) #True면 이 텐서에 대해 수행되는 연산들이 기록됨
float_32_tensor.shape, float_32_tensor.dtype, float_32_tensor.device
```

*requires_grad : 자동미분(autograd)을 켤지 말지 정하는 옵션 <br>
    -False면 연산 추적x. 기울기 계산 안 함

Pytorch에서 흔히 접하는 문제 <br>
1. 텐서 형태 일치하지 않음
2. 데이터 유형 일치하지 않음

# Matrix multiplication(행렬곱) 

@ :  파이썬에서 행렬곱셈 나타내는 기호

- 주요 규칙
1. 내부 차원 일치 <br>
(3, 2) @ (3, 2) -> 작동 x  <br>
(2, 3) @ (3, 2) -> 작동 <br>

2. 결과 행렬은 외부 차원 형태 가짐 <br>
(2, 3) @ (3, 2) -> (2, 2)



```python
import torch 
tensor =  torch.tensor( [1, 2, 3] )
tensor.shape
```

요쇼별 곱셈 (element-wise multiplication) : tensor * tensor --> [1*1, 2*2, 3*3]=[1, 4, 9]  <br>
행렬 곱셈 (matrix multiplication): tensor.matmul(tensor) --> [1*1 + 2*2 + 3*3]=[14]

```python
tensor*tensor
```

```python
torch.matmul(tensor, tensor)
```

```python
tensor @ tensor
```

# 딥러닝에서 가장 흔한 오류-형태 오류(모양 불일치)

- 전치 (주어진 텐서의 치원 바꾸는 것)
1) torch.transpose(input, dim0, dim1) - input(전치하려는 텐서), dim0 & dim1 (교환할 차원)
2) tensor.T : 텐서의 마지막 두 차원을 변경


```python
tensor_A = torch.tensor([[1, 2],
                         [3, 4],
                         [5, 6]], dtype=torch.float32)
tensor_B = torch.tensor([[7, 10],
                         [8, 11], 
                         [9, 12]], dtype=torch.float32)
```

```python
print(tensor_A)
print(tensor_B.T)
```

```python
print(f"Original shapes: tensor_A = {tensor_A.shape}, tensor_B = {tensor_B.shape}\n")   #원래 형태 
print(f"New shapes: tensor_A = {tensor_A.shape} (same as above), tensor_B.T = {tensor_B.T.shape}\n") # 바뀐 후 
print(f"Multiplying: {tensor_A.shape} * {tensor_B.T.shape} <- inner dimensions match\n") #내부 차원이 일치해 행렬곱 가능
print("Output:\n")
output = torch.matmul(tensor_A, tensor_B.T)
print(output) 
print(f"\nOutput shape: {output.shape}")
```

- torch.mm( ) = torch.matmul ( )
